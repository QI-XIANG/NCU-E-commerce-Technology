## 1.載入 customer_churn.csv 為 pandas DataFrame 格式

```python
import pandas as pd 
customerChurn = pd.read_csv('customer_churn.csv')
customerChurn
```

## 2.列出資料筆數、屬性數量、每個欄位的空值個數以及各類別(target)的資料筆數

## 3.資料集中有部分資料重覆出現，請刪除重覆多餘的資料 (僅保留一筆)，並列出剩餘的資料筆數。

```python
customerChurn = customerChurn.drop_duplicates()
```

```python
customerChurn
```

```python
customerChurn.info()
```

## 4.填補空值及其他資料前處理 (e.g., feature encoding)

```python
## 刪掉 CustomerID
customerChurn = customerChurn.drop(['CustomerID'],axis=1)
customerChurn
```

```python
## 填補缺失值 都用 mean 或 mode 填
customerChurn['Tenure'] = customerChurn['Tenure'].fillna(customerChurn['Tenure'].mean())
customerChurn['WarehouseToHome'] = customerChurn['WarehouseToHome'].fillna(customerChurn['WarehouseToHome'].mean())
customerChurn['HourSpendOnApp'] = customerChurn['HourSpendOnApp'].fillna(customerChurn['HourSpendOnApp'].mean())
customerChurn['OrderAmountHikeFromlastYear'] = customerChurn['OrderAmountHikeFromlastYear'].fillna(customerChurn['OrderAmountHikeFromlastYear'].mean())
customerChurn['DaySinceLastOrder'] = customerChurn['DaySinceLastOrder'].fillna(customerChurn['DaySinceLastOrder'].mean())
customerChurn['CouponUsed'] = customerChurn['CouponUsed'].fillna(customerChurn['CouponUsed'].mean())
customerChurn['OrderCount'] = customerChurn['OrderCount'].fillna(customerChurn['OrderCount'].mean())
```

```python
customerChurn
```

```python
## nominal to numeric
dummyColumn = pd.get_dummies(customerChurn[['PreferredLoginDevice','PreferredPaymentMode','Gender','PreferedOrderCat','MaritalStatus']])
dff = pd.DataFrame(dummyColumn)
customerChurn = pd.concat([customerChurn, dff],axis=1)
customerChurn = customerChurn.drop(['PreferredLoginDevice','PreferredPaymentMode','Gender','PreferedOrderCat','MaritalStatus'], axis=1)
customerChurn
```

```python
customerChurn.info()
```

```python
## 將 churn 換成 nominal
customerChurn['Churn'] = customerChurn['Churn'].astype('string')
customerChurn.info()
```

## 5.將資料集劃分為訓練集與測試集

```python
from sklearn.model_selection import train_test_split
X = customerChurn.drop(['Churn'], axis=1)
y = customerChurn['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state=12)
```

## 6.訓練與測試 SVM、Logistic Regression、Decision Tree 模型

### SVM

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
from sklearn.metrics import accuracy_score
print("SVM Accuracy:",accuracy_score(y_test, y_pred))
```

### Logistic Regression

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
clf2 = LogisticRegression(random_state=0).fit(X_train, y_train)
y_pred2 =clf2.predict(X_test)

print("Logistic Regression Accuracy:",accuracy_score(y_test, y_pred2))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(random_state=0, max_depth=5)
decision_tree = decision_tree.fit(X_train, y_train)
y_pred3 = decision_tree.predict(X_test)

print("Decision Tree Accuracy:",accuracy_score(y_test, y_pred3))
```
